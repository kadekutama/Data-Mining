### Load CSV File
The code below is used to read data from CSV and display the contents of the data frame.

```python
import pandas as pd

df = pd.read_csv("credit-train.csv")
df
```

### Replacing Empty Values on the Data Frame
Before the data frame is processed, null value must be replaced by a value corresponding to the column data type. This is done to avoid errors due to a null value.

```python
df['Loan_ID'].fillna('Unknown', inplace = True)
df['Gender'].fillna('Unknown', inplace = True)
df['Married'].fillna('Unknown', inplace = True)
df['Dependents'].fillna(0, inplace = True)
df['Education'].fillna('Unknown', inplace = True)
df['Self_Employed'].fillna('Unknown', inplace = True)
df['ApplicantIncome'].fillna(0, inplace = True)
df['CoapplicantIncome'].fillna(0, inplace = True)
df['LoanAmount'].fillna(0, inplace = True)
df['Loan_Amount_Term'].fillna(0, inplace = True)
df['Credit_History'].fillna(0, inplace = True)
df['Property_Area'].fillna('Unknown', inplace = True)
df['Loan_Status'].fillna('Unknown', inplace = True)

df.head(5)
```

### Changing Mismatch Format Values
In the column data frame "dependents", there are values that do not fit the format, which is "3+". The value of "3+" needs to be converted into a "3" to avoid errors due to the type of data that is not the same in a column.

```python
df['Dependents'].replace(['3+'],[3], inplace = True)

df.head(10)
```

### Changing the Data Type Each Column
CSV file data types are generally in the form of string, so that the data type of each column must be changed in accordance with the type of original, or it could be appropriate. Conversion of data types will be particularly important if the operation is done the calculations.

```python
df['Loan_ID'] = df['Loan_ID'].astype('str')
df['Gender'] = df['Gender'].astype('str')
df['Married'] = df['Married'].astype('str')
df['Dependents'] = df['Dependents'].astype('int')
df['Education'] = df['Education'].astype('str')
df['Self_Employed'] = df['Self_Employed'].astype('str')
df['ApplicantIncome'] = df['ApplicantIncome'].astype('int')
df['CoapplicantIncome'] = df['CoapplicantIncome'].astype('int')
df['LoanAmount'] = df['LoanAmount'].astype('int')
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].astype('int')
df['Credit_History'] = df['Credit_History'].astype('int')
df['Property_Area'] = df['Property_Area'].astype('str')
df['Loan_Status'] = df['Loan_Status'].astype('str')

df.head(5)
```

### Delete Duplicate Data
It often happens duplication of data in the data frame, so that the same data must be deleted to avoid inconsistencies.

```python
print(df.duplicated())
# Tidak ada data yang terduplikasi. Kode di bawah hanya contoh.
df.drop_duplicates(['Property_Area','Loan_Status'])
```

### Data Set Transpose
Oftentimes we just need some data only from a data frame. We can customize what data will be shown for ease of processing.

```python
#tp = df.pivot(index = 'Loan_ID', columns='Property_Area', values = 'LoanAmount')
tp = df.pivot(index = 'Loan_ID', columns='Property_Area', values = 'Loan_Status')
tp.fillna('-', inplace = True)
tp
```

### Data Sorting
Data sorting is performed to determine priorities. Ordering can be done with some parameters, not just one column reference.

```python
df.sort_values(['Self_Employed','ApplicantIncome'],ascending = [True, False])
```

### Creating Data Plot
Data visualization can help us to better understand the data.

```python
import matplotlib.pyplot as plt

fig = plt.figure()
hst = fig.add_subplot(1,1,1)
hst.hist(df['Loan_Amount_Term'])
plt.title('Loan Amount Term Distribution')
plt.xlabel('Loan Amount Term')
plt.ylabel('Total Loan Amount Term')
plt.show()
```

```python
fig = plt.figure()
sct = fig.add_subplot(1,1,1)
sct.scatter(df['Loan_Amount_Term'],df['LoanAmount'])
plt.title('Loan Amount and Loan Amount Term Distribution')
plt.xlabel('Loan Amount Term')
plt.ylabel('Loan Amount')
plt.show()
```

```python
import seaborn as sb

sb.boxplot([df['Loan_Amount_Term'],df['LoanAmount']])
#sb.despine()
plt.show()
```

### Variable Grouping and Calculations
Variable grouping can be used for various purposes, such as creating a frequency table, as well as counting the number, the average value, maximum/minimum value, standard deviation, etc..

```python
freq = df.groupby(['Property_Area','Loan_Status'])
print(freq.size())
freq.describe()
```
